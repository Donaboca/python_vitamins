<h4>Lopullinen Data Frame, johon valitaan halutut nimikkeet. Myös vanhasta idenksoinnista luovutaan, ettei se seuraa mukana. Lisäksi järjestetään DF aakkosjärjestykseen.</h4>

```python
healthy_df = ndf.loc[ndf["Livsmedelsnamn"].isin(product)]
```

```python
healthy_df.reset_index(drop=True)
```

```python
healthy_df = healthy_df.sort_values(by=["Livsmedelsnamn"])
healthy_df = healthy_df.reset_index(drop=True)
healthy_df
```

<h4>Vaihdetaan Livsmedelsnamn sarakkeen nimi. Sen jälkeen muutetaan indeksointi Product sarakkeelle.</h4>

```python
healthy_df = healthy_df.rename(columns={"Livsmedelsnamn": "Product"})
```

```python
healthy_df = healthy_df.set_index("Product")
```
