```python
import pymongo
import pandas as pd
from pymongo import MongoClient
client = MongoClient("localhost", 27017)
# Luodaan yhteys tietokantaan - jos tämän nimistä tietokantaa ei ole, se luodaan samalla.
db = client["python_vitamins"]
# Kokoelma
healthy_products = db["Healthy_products"]
```

<h4>Luetaan data Excelistä data frameen, resetoidaan indeksi ja muunnetaan data frame sanakirjaksi. Sen jälkeen viedään sanakirja MongoDB:een. Tieto on tämän jälkeen MongoDB:ssa yhtenä arrayna.</h4>

```python
df_to_mongo = pd.read_excel("C:/Users/mynttja/GIT/Python_vitamins/nutrition_ready.xlsx")
df_to_mongo.reset_index(inplace=True)
```

```python
df_dict = df_to_mongo.to_dict("records")
```

```python
healthy_products.insert_many(df_dict)
```

<h4>Haetaan testin vuoksi tietokannan ensimmäinen dokumentti.</h4>

```python
healthy_products.find_one()
```

<h4>Ja testin vuoksi tietokannan kokoelma data frameksi</h4>

```python
healthy_products_cursor = healthy_products.find()
test_df = pd.DataFrame(healthy_products_cursor)
test_df.set_index("Product", inplace=True)
test_df.drop(["_id", "index"], axis=1, inplace=True)
test_df
```
